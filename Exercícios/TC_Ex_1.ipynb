## Questão 5

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# =========================
# Sinais definidos
# =========================

x = np.array([2, -1, 0, 1])
y = np.array([0, 1, 2, -1])

Nx = len(x)
Ny = len(y)

# Valores possíveis de atraso
lags = np.arange(-(Nx-1), Ny)

# =========================
# Correlação cruzada
# Rxy[l] = sum x[n+l] y[n]
# =========================

def cross_correlation(x, y, lags):
    r = []

    for l in lags:
        s = 0

        for n in range(len(y)):

            xn_index = n + l

            if 0 <= xn_index < len(x):
                s += x[xn_index] * y[n]

        r.append(s)

    return np.array(r)

Rxy = cross_correlation(x, y, lags)

# =========================
# Limites FIXOS dos gráficos
# =========================

signal_min = min(np.min(x), np.min(y)) - 1
signal_max = max(np.max(x), np.max(y)) + 1

corr_min = np.min(Rxy) - 1
corr_max = np.max(Rxy) + 1

# =========================
# Função de visualização
# =========================

def plot_corr(l):

    fig, ax = plt.subplots(2, 1, figsize=(12, 8))

    # ==================================================
    # GRÁFICO 1 -> sinais sobrepostos
    # ==================================================

    n_x = np.arange(len(x))
    n_y = np.arange(len(y))

    # deslocamento de x[n+l]
    shifted_x_axis = n_x - l

    ax[0].stem(
        shifted_x_axis,
        x,
        linefmt='C0-',
        markerfmt='C0o',
        basefmt='k-',
        label='x[n+l]'
    )

    ax[0].stem(
        n_y,
        y,
        linefmt='C1-',
        markerfmt='C1s',
        basefmt='k-',
        label='y[n]'
    )

    ax[0].set_title(f'Sinais Sobrepostos para l = {l}')
    ax[0].set_xlabel('n')
    ax[0].set_ylabel('Amplitude')

    # FIXANDO eixos
    ax[0].set_xlim(-5, 8)
    ax[0].set_ylim(signal_min, signal_max)

    ax[0].grid(True)
    ax[0].legend()

    # ==================================================
    # GRÁFICO 2 -> correlação cruzada
    # ==================================================

    markerline, stemlines, baseline = ax[1].stem(
        lags,
        Rxy,
        linefmt='gray',
        markerfmt='ko',
        basefmt='k-'
    )

    # destacar lag atual
    idx = np.where(lags == l)[0][0]

    ax[1].stem(
        [lags[idx]],
        [Rxy[idx]],
        linefmt='r-',
        markerfmt='ro',
        basefmt='k-'
    )

    ax[1].set_title('Correlação Cruzada $R_{xy}[l]$')
    ax[1].set_xlabel('Atraso l')
    ax[1].set_ylabel('Correlação')

    # FIXANDO eixos
    ax[1].set_xlim(lags[0]-1, lags[-1]+1)
    ax[1].set_ylim(corr_min, corr_max)

    ax[1].grid(True)

    plt.tight_layout()
    plt.show()

# =========================
# Slider interativo
# =========================

interact(
    plot_corr,
    l=IntSlider(
        min=lags[0],
        max=lags[-1],
        step=1,
        value=1,
        description='Atraso l'
    )
)
```

## Questão 6

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# ==========================================================
# QUESTÃO 6
# ==========================================================

# Pulso transmitido
x = np.array([1, 2, 1])

# Sinal recebido com eco
y = np.array([0, 0, 0, 1, 2, 1, 0])

nx = np.arange(len(x))
ny = np.arange(len(y))

# ==========================================================
# Correlação cruzada
# Definição:
#
# Rxy[l] = Σ x[n+l] y[n]
#
# O pico ocorre em l = -3
# ==========================================================

lags = np.arange(-(len(y)-1), len(x))

Rxy = []

for l in lags:

    soma = 0

    for n in range(len(y)):

        idx = n + l

        if 0 <= idx < len(x):
            soma += x[idx] * y[n]

    Rxy.append(soma)

Rxy = np.array(Rxy)

# ==========================================================
# LIMITES FIXOS DOS EIXOS
# ==========================================================

xmin_signal = -1
xmax_signal = len(y) + 1

ymin_signal = min(np.min(x), np.min(y)) - 1
ymax_signal = max(np.max(x), np.max(y)) + 1

xmin_corr = lags[0] - 1
xmax_corr = lags[-1] + 1

ymin_corr = np.min(Rxy) - 1
ymax_corr = np.max(Rxy) + 1

# ==========================================================
# FUNÇÃO DE VISUALIZAÇÃO
# ==========================================================

def visualizar_correlacao(l):

    fig, ax = plt.subplots(2, 1, figsize=(13, 8))

    # ======================================================
    # PRIMEIRO GRÁFICO
    # ======================================================

    # x deslocado de acordo com o lag
    x_shift_pos = nx - l

    # Sinal recebido
    ax[0].stem(
        ny,
        y,
        linefmt='C1-',
        markerfmt='C1o',
        basefmt='k-',
        label='y[n]'
    )

    # Pulso deslocado
    ax[0].stem(
        x_shift_pos,
        x,
        linefmt='C0-',
        markerfmt='C0s',
        basefmt='k-',
        label=f'x[n+l], l={l}'
    )

    ax[0].set_title('Alinhamento dos sinais')
    ax[0].set_xlabel('n')
    ax[0].set_ylabel('Amplitude')

    # EIXOS FIXOS
    ax[0].set_xlim(xmin_signal, xmax_signal)
    ax[0].set_ylim(ymin_signal, ymax_signal)

    ax[0].grid(True)
    ax[0].legend()

    # ======================================================
    # SEGUNDO GRÁFICO
    # ======================================================

    ax[1].stem(
        lags,
        Rxy,
        linefmt='gray',
        markerfmt='ko',
        basefmt='k-'
    )

    # Destacar lag atual
    idx_lag = np.where(lags == l)[0][0]

    ax[1].stem(
        [l],
        [Rxy[idx_lag]],
        linefmt='r-',
        markerfmt='ro',
        basefmt='k-'
    )

    ax[1].set_title(r'Correlação Cruzada $R_{xy}[l]$')
    ax[1].set_xlabel('Lag (l)')
    ax[1].set_ylabel(r'$R_{xy}[l]$')

    # EIXOS FIXOS
    ax[1].set_xlim(xmin_corr, xmax_corr)
    ax[1].set_ylim(ymin_corr, ymax_corr)

    ax[1].grid(True)

    plt.tight_layout()
    plt.show()

    print(f'Lag selecionado: l = {l}')
    print(f'Rxy[{l}] = {Rxy[idx_lag]}')


# ==========================================================
# SLIDER INTERATIVO
# ==========================================================

interact(
    visualizar_correlacao,
    l=IntSlider(
        min=lags[0],
        max=lags[-1],
        step=1,
        value=-3,   # pico correto
        description='Lag'
    )
)
```
